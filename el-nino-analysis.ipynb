# Getting to know the dataset


## This dataset contains information about factors that lead to the El Nino Southern Oscillations.



### El Niño–Southern Oscillation (ENSO) is a global climate phenomenon that emerges from variation in winds and sea surface temperatures over the tropical Pacific Ocean. This dataset contains featuresused for El Nino prediction.

```python
df_original = pd.read_csv('/kaggle/input/datasets/organizations/uciml/el-nino-dataset/elnino.csv')
```

## This dataset contains values from 1980 to 1998. But only the last two digits of the years is mentioned in the data. So, I added 1900 to each datapoint in the year column.

```python
df_original["Year"]
```

```python
df_original["Year"] = df_original["Year"]+1900
```
